# Activation Functions and Their Derivatives
Sigmoid, Tanh, ReLU, ELU, SELU in range (-10, 10)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(-10, 10, 1000)
```

```python
print(x)
print(x.shape)
```

```python


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def selu(x, lambda_=1.0507, alpha=1.67326):
    return lambda_ * np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python

def d_sigmoid(x):
    s = sigmoid(x)
    return s * (1 - s)

def d_tanh(x):
    return 1 - np.tanh(x)**2

def d_relu(x):
    return np.where(x > 0, 1, 0)

def d_elu(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))

def d_selu(x, lambda_=1.0507, alpha=1.67326):
    return lambda_ * np.where(x > 0, 1, alpha * np.exp(x))
```

```python
functions = [
    ("Sigmoid", sigmoid, d_sigmoid),
    ("Tanh", tanh, d_tanh),
    ("ReLU", relu, d_relu),
    ("ELU", elu, d_elu),
    ("SELU", selu, d_selu),
]
```

```python




for name, func, dfunc in functions:
    plt.figure()
    plt.plot(x, func(x))
    plt.title(f"{name} Function")
    plt.xlabel("x")
    plt.ylabel(f"{name}(x)")
    plt.grid(True)
    plt.show()

    plt.figure()
    plt.plot(x, dfunc(x))
    plt.title(f"{name} Derivative")
    plt.xlabel("x")
    plt.ylabel(f"{name}'(x)")
    plt.grid(True)
    plt.show()
```
